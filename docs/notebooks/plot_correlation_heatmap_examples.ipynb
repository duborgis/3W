
# Correlation Heatmap Demonstration

This notebook demonstrates how to use the `PlotCorrelationHeatmap.correlation_heatmap` function in various scenarios.

The function accepts:
- A DataFrame of time series (`df_of_series`)
- Additional keyword arguments (`**kwargs`) for customization (e.g., `title`, `figsize`, `annot`, `cmap`, `ax`, etc.)


## Using mock data as example

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ThreeWToolkit.data_visualization import PlotCorrelationHeatmap
```

### Basic Usage

```python
df = pd.DataFrame(np.random.rand(100, 5), columns=["A", "B", "C", "D", "E"])
fig = PlotCorrelationHeatmap.correlation_heatmap(df)
fig.show()
```

### Custom Title and Annotations

```python

fig = PlotCorrelationHeatmap.correlation_heatmap(
    df,
    title="Annotated Heatmap",
    annot=True,
    cmap="viridis",
    fmt=".2f"
)
fig.show()
```

### Custom Figure Size and Color Map

```python

fig = PlotCorrelationHeatmap.correlation_heatmap(
    df,
    title="Larger Heatmap with 'coolwarm'",
    figsize=(14, 10),
    cmap="coolwarm",
    annot=True
)
fig.show()
```

### Multiple Heatmaps in Subplots

```python

df2 = pd.DataFrame(np.random.rand(100, 4), columns=["W", "X", "Y", "Z"])

fig, axs = plt.subplots(1, 2, figsize=(16, 6))
PlotCorrelationHeatmap.correlation_heatmap(df, ax=axs[0], title="Dataset 1", annot=True)
PlotCorrelationHeatmap.correlation_heatmap(df2, ax=axs[1], title="Dataset 2", cmap="crest", annot=True)
plt.tight_layout()
plt.show()
```

## Using 3W data as example

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ThreeWToolkit.data_visualization import PlotCorrelationHeatmap
```

Loading the 3W dataset

```python
from ThreeWToolkit.core.base_dataset import ParquetDatasetConfig
from ThreeWToolkit.dataset import ParquetDataset
from pathlib import Path

# define path
dataset_path = Path("./data/3w_dataset/")

# create and load
ds_config = ParquetDatasetConfig(path=dataset_path, clean_data=False, download=True)
ds = ParquetDataset(ds_config)
len(ds)
```

```python
sig = ds[123]["signal"].copy()
```

### Basic Usage

```python
features = ['P-ANULAR', 'P-JUS-CKGL', 'P-MON-CKP', 'P-TPT', 'T-JUS-CKP', 'T-TPT']
subset = sig[features]

fig = PlotCorrelationHeatmap.correlation_heatmap(
    df_of_series=subset,
    title="Correlation Heatmap of Selected Features"
)
fig.show()
```

### Custom Title and Annotations

```python
features = ['P-ANULAR', 'P-JUS-CKGL', 'P-MON-CKP', 'P-TPT', 'T-JUS-CKP', 'T-TPT']
subset = sig[features]

fig = PlotCorrelationHeatmap.correlation_heatmap(
    subset,
    title="Annotated Heatmap",
    annot=True,
    cmap="viridis",
    fmt=".2f"
)
fig.show()
```

### Custom Figure Size and Color Map

```python
features = ['P-ANULAR', 'P-JUS-CKGL', 'P-MON-CKP', 'P-TPT', 'T-JUS-CKP', 'T-TPT']
subset = sig[features]

fig = PlotCorrelationHeatmap.correlation_heatmap(
    subset,
    title="Larger Heatmap with 'coolwarm'",
    figsize=(14, 10),
    cmap="coolwarm",
    annot=True
)
fig.show()
```

### Multiple Heatmaps in Subplots

```python
features_1 = ['P-ANULAR', 'P-JUS-CKGL', 'P-MON-CKP', 'P-TPT', 'T-JUS-CKP', 'T-TPT']
subset_1 = sig[features_1]

features_2 = ['T-TPT', 'P-JUS-CKGL', 'P-TPT', 'P-MON-CKP', 'T-JUS-CKP', 'P-ANULAR']
subset_2 = sig[features_2]

fig, axs = plt.subplots(1, 2, figsize=(16, 6))
PlotCorrelationHeatmap.correlation_heatmap(subset_1, ax=axs[0], title="Dataset 1", annot=True)
PlotCorrelationHeatmap.correlation_heatmap(subset_2, ax=axs[1], title="Dataset 2", cmap="crest", annot=True)
plt.tight_layout()
plt.show()
```
